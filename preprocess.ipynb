```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv("goemotions_1.csv")
df_2 = pd.read_csv("goemotions_2.csv")
df_3 = pd.read_csv("goemotions_3.csv")

df_train = pd.read_csv('train.txt', header =None, sep =';', names = ['Input','Sentiment'], encoding='utf-8')
df_test = pd.read_csv('test.txt', header = None, sep =';', names = ['Input','Sentiment'],encoding='utf-8')
df_val = pd.read_csv('val.txt',header=None,sep=';', names=['Input','Sentiment'], encoding='utf-8')

df_train['Sentiment'] = df_train['Sentiment'].replace('love', 'trust')
df_test['Sentiment'] = df_test['Sentiment'].replace('love', 'trust')
df_val['Sentiment'] = df_test['Sentiment'].replace('love', 'trust')


filtered_df1 = df_1[~df_1['text'].str.contains("NAME")]
filtered_df2 = df_2[~df_2['text'].str.contains("NAME")]
filtered_df3 = df_3[~df_3['text'].str.contains("NAME")]

pd.options.display.max_colwidth = 200
df_train
# df_1
```

```python
fear_df_1 = filtered_df1[filtered_df1['fear'] == 1]
disgust_df_1 = filtered_df1[filtered_df1['disgust'] == 1]
anger_df_1 = filtered_df1[filtered_df1['anger'] == 1]
surprise_df_1 = filtered_df1[filtered_df1['surprise'] == 1]

fear_df_2 = filtered_df2[filtered_df2['fear'] == 1]
disgust_df_2 = filtered_df2[filtered_df2['disgust'] == 1]
surprise_df_2 = filtered_df2[filtered_df2['surprise'] == 1]
# anticipation_df_2 = filtered_df2[filtered_df2['optimism'] == 1]

fear_df_3 = filtered_df3[filtered_df3['fear'] == 1]
disgust_df_3 = filtered_df3[filtered_df3['disgust'] == 1]
love_df_3 = filtered_df3[filtered_df3['love'] == 1]

# selected_cols_anticipation_1 = ['text', 'optimism']
# filtered_d_anticipation_1 = anticipation_df_2[selected_cols_anticipation_1]
# filtered_d_anticipation_1 = filtered_d_anticipation_1.rename(columns={'text': 'Input', 'optimism': 'Sentiment'})
# filtered_d_anticipation_1['Sentiment'] = filtered_d_anticipation_1['Sentiment'].replace(1, 'anticipation')

selected_cols_anger_1 = ['text', 'anger']
filtered_d_anger_1 = anger_df_1[selected_cols_anger_1]
filtered_d_anger_1 = filtered_d_anger_1.rename(columns={'text': 'Input', 'anger': 'Sentiment'})
filtered_d_anger_1['Sentiment'] = filtered_d_anger_1['Sentiment'].replace(1, 'anger')

selected_cols_surprise_1 = ['text', 'surprise']
filtered_d_surprise_1 = surprise_df_1[selected_cols_surprise_1]
filtered_d_surprise_1 = filtered_d_surprise_1.rename(columns={'text': 'Input', 'surprise': 'Sentiment'})
filtered_d_surprise_1['Sentiment'] = filtered_d_surprise_1['Sentiment'].replace(1, 'surprise')

selected_cols_surprise_1 = ['text', 'surprise']
filtered_d_surprise_2 = surprise_df_2[selected_cols_surprise_1]
filtered_d_surprise_2 = filtered_d_surprise_2.rename(columns={'text': 'Input', 'surprise': 'Sentiment'})
filtered_d_surprise_2['Sentiment'] = filtered_d_surprise_2['Sentiment'].replace(1, 'surprise')

selected_cols_love_1 = ['text', 'love']
filtered_d_love_1 = love_df_3[selected_cols_love_1]
filtered_d_love_1 = filtered_d_love_1.rename(columns={'text': 'Input', 'love': 'Sentiment'})
filtered_d_love_1['Sentiment'] = filtered_d_love_1['Sentiment'].replace(1, 'trust')

selected_cols_fear_1 = ['text', 'fear']
filtered_d_fear_1 = fear_df_1[selected_cols_fear_1]
filtered_d_fear_1 = filtered_d_fear_1.rename(columns={'text': 'Input', 'fear': 'Sentiment'})
filtered_d_fear_1['Sentiment'] = filtered_d_fear_1['Sentiment'].replace(1, 'fear')

selected_cols_fear_2 = ['text', 'fear']
filtered_d_fear_2 = fear_df_2[selected_cols_fear_2]
filtered_d_fear_2 = filtered_d_fear_2.rename(columns={'text': 'Input', 'fear': 'Sentiment'})
filtered_d_fear_2['Sentiment'] = filtered_d_fear_2['Sentiment'].replace(1, 'fear')

selected_cols_fear_3 = ['text', 'fear']
filtered_d_fear_3 = fear_df_3[selected_cols_fear_3]
filtered_d_fear_3 = filtered_d_fear_3.rename(columns={'text': 'Input', 'fear': 'Sentiment'})
filtered_d_fear_3['Sentiment'] = filtered_d_fear_3['Sentiment'].replace(1, 'fear')

selected_cols_disgust = ['text', 'disgust']
filtered_d_disgust_1 = disgust_df_1[selected_cols_disgust]
filtered_d_disgust_1 = filtered_d_disgust_1.rename(columns={'text': 'Input', 'disgust': 'Sentiment'})
filtered_d_disgust_1['Sentiment'] = filtered_d_disgust_1['Sentiment'].replace(1, 'disgust')

selected_cols_disgust = ['text', 'disgust']
filtered_d_disgust_2 = disgust_df_2[selected_cols_disgust]
filtered_d_disgust_2 = filtered_d_disgust_2.rename(columns={'text': 'Input', 'disgust': 'Sentiment'})
filtered_d_disgust_2['Sentiment'] = filtered_d_disgust_2['Sentiment'].replace(1, 'disgust')


selected_cols_disgust = ['text', 'disgust']
filtered_d_disgust_3 = disgust_df_3[selected_cols_disgust]
filtered_d_disgust_3 = filtered_d_disgust_3.rename(columns={'text': 'Input', 'disgust': 'Sentiment'})
filtered_d_disgust_3['Sentiment'] = filtered_d_disgust_3['Sentiment'].replace(1, 'disgust')

fear_data = pd.DataFrame(columns=['Input', 'Sentiment'])
fear_data = pd.concat([fear_data, filtered_d_fear_1], ignore_index=True)
fear_data = pd.concat([fear_data, filtered_d_fear_2], ignore_index=True)
fear_data = pd.concat([fear_data, filtered_d_fear_3], ignore_index=True)

disgust_data = pd.DataFrame(columns=['Input', 'Sentiment'])
disgust_data = pd.concat([disgust_data, filtered_d_disgust_1], ignore_index=True)
disgust_data = pd.concat([disgust_data, filtered_d_disgust_2], ignore_index=True)
disgust_data = pd.concat([disgust_data, filtered_d_disgust_3], ignore_index=True)

anger_data = pd.DataFrame(columns=['Input', 'Sentiment'])
anger_data = pd.concat([anger_data, filtered_d_anger_1], ignore_index=True)

surprise_data = pd.DataFrame(columns=['Input', 'Sentiment'])
surprise_data = pd.concat([surprise_data, filtered_d_surprise_1], ignore_index=True)
surprise_data = pd.concat([surprise_data, filtered_d_surprise_2], ignore_index=True)

love_data = pd.DataFrame(columns=['Input', 'Sentiment'])
love_data = pd.concat([love_data, filtered_d_love_1], ignore_index=True)

# print(love_data)

# anticipation_data = pd.DataFrame(columns=['Input', 'Sentiment'])
# anticipation_data = pd.concat([anticipation_data, filtered_d_anticipation_1], ignore_index=True)

num_rows = fear_data.shape[0]
print(num_rows)

num_row1 = disgust_data.shape[0]
print(num_row1)
num_row2 = anger_data.shape[0]
print(num_row2)

num_row3 = surprise_data.shape[0]
print(num_row3)

num_row4 = love_data.shape[0]
print(num_row4)

# num_row5 = anticipation_data.shape[0]
# print(num_row5)




# final_df = pd.concat([final_df, fear_df_1], ignore_index=True)




# num_rows = disgust_df.shape[0]
# num_rows1 = df.shape[0]
# print("Number of rows in filtered_df:", num_rows)
# print("Number of rows in filtered_df:", num_rows1)
```

```python
# Assuming you have additional data for "fear" and "disgust" in separate DataFrames
# Sample a portion of the additional data for each class
fear_data = fear_data.sample(n=2700, random_state=42)
disgust_data = disgust_data.sample(n=4600, random_state=42)
anger_data = anger_data.sample(n=2200, random_state=42)
surprise_data = surprise_data.sample(n=3000, random_state=42)
love_data = love_data.sample(n=2200, random_state=42)
# anticipation_data = anticipation_data.sample(n=2500, random_state=42)




# # Concatenate the sampled data with your train, test, and validate DataFrames
# train_df = pd.concat([df_train, fear_data, disgust_data, anger_data, surprise_data, love_data, anticipation_data], ignore_index=True)
# test_df = pd.concat([df_test, fear_data, disgust_data, anger_data, surprise_data, love_data, anticipation_data], ignore_index=True)
# validate_df = pd.concat([df_val, fear_data, disgust_data, anger_data, surprise_data, love_data, anticipation_data], ignore_index=True)

# Concatenate the sampled data with your train, test, and validate DataFrames
train_df = pd.concat([df_train, fear_data, disgust_data, anger_data, surprise_data, love_data], ignore_index=True)
test_df = pd.concat([df_test, fear_data, disgust_data, anger_data, surprise_data, love_data], ignore_index=True)
validate_df = pd.concat([df_val, fear_data, disgust_data, anger_data, surprise_data, love_data], ignore_index=True)

# Shuffle the concatenated DataFrames to randomize the order of rows
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)
test_df = test_df.sample(frac=1, random_state=42).reset_index(drop=True)
validate_df = validate_df.sample(frac=1, random_state=42).reset_index(drop=True)


# train_df['Input'] = train_df['Input'].replace('love', 'trust')
# test_df['Input'] = test_df['Input'].replace('love', 'trust')
# validate_df['Input'] = validate_df['Input'].replace('love', 'trust')

# train_df['Input'] = train_df['Input'].replace('admiration', 'anticipation')
# test_df['Input'] = test_df['Input'].replace('admiration', 'anticipation')
# validate_df['Input'] = validate_df['Input'].replace('admiration', 'anticipation')

# Save the updated train, test, and validate DataFrames to CSV files
train_df.to_csv('final_train.csv', index=False)
test_df.to_csv('final_test.csv', index=False)
validate_df.to_csv('final_validate.csv', index=False)

```
