```python
import pandas as pd
from difflib import SequenceMatcher

# Read the two CSV files into DataFrames
df1 = pd.read_csv('final_train.csv')
df2 = pd.read_csv('final_test.csv')

# Define a function to calculate similarity between two rows
def similarity(row1, row2):
    # You can use different similarity metrics here
    # In this example, we use the SequenceMatcher ratio
    return SequenceMatcher(None, row1, row2).ratio()

# Iterate through rows in df1 and df2
for index1, row1 in df1.iterrows():
    for index2, row2 in df2.iterrows():
        # Compare the similarity between rows
        similarity_score = similarity(row1['Input'], row2['Input'])

        # Set a threshold for similarity (adjust as needed)
        similarity_threshold = 0.8  # Example threshold

        if similarity_score >= similarity_threshold:
            print(f"Similar rows found: Row {index1} in file1 and Row {index2} in file2")

# Note: This example uses the SequenceMatcher ratio for similarity,
# but you can use other similarity measures based on your requirements.
```
